```python
import pandas as pd
from chatbot.AN_Knowledge_Base import AN_KB_Manager
```

```python
# import data
an_kb = AN_KB_Manager()
df_qna = pd.DataFrame(an_kb.qna_manager.get_unanswered_questions())

df_qna.head()
```

```python
# Define color formatting for the "Status" column
def status_color_formatter(status):
    if status == "answered":
        return "background-color: #d4edda; color: #155724;" 
    elif status == "unanswered":
        return "background-color: #f8d7da; color: #721c24;" 
    return ""  # default styling
```

```python
df_qna.style.applymap(status_color_formatter, subset=pd.IndexSlice[:, ['status']])
```
